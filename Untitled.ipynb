```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import datetime
import numpy as np

def CSI(a1, a2, a3, a4):
    vals = [0, 11/4, 11/2, 10]
    return a1 * vals[0] + a2 * vals[1] + a3 * vals[2] + a4 * vals[3] 

fig, ax = plt.subplots()
index = np.arange(4)

x = [3, 3, 3, 3, 2, 2, 3, 3, 2]
y = [2, 2, 2, 2, 1, 1, 2, 2]
counts_x = [0, 0, 3, 6]
counts_y = [0, 2, 6, 0]
bar_colors = ['#c71d1d', '#e28e8e', '#879caf', '#1b617b']

bar_width = 0.35

GUI = ax.bar(index, counts_x, bar_width, label="GUI") #left=proportion_strongdisagree+proportion_disagree+proportion_neutral+proportion_agree)
SD = ax.bar(index+bar_width, counts_y, bar_width, label="Prompt")

ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.set_ylabel('Number Answered')
ax.set_title('I enjoyed using the system or tool')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(['Strongly Disagree', 'Disagree', 'Agree', 'Strongly Agree'])
ax.legend()

plt.show()
```
